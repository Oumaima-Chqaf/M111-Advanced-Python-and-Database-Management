## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt
graph=pd.read_csv("../data/company_sales_data.csv")
plt.figure(figsize=(20,6))
axis1=graph["total_profit"]
axis0=graph["month_number"]
plt.plot(axis0,axis1,scalex=True)
plt.xticks(range(1,13))
plt.xlabel("Month number")
plt.ylabel("Profit in dollar")
plt.title("Company profit per month")
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
circledata  = [graph['facecream'].sum(), graph['facewash'].sum(), 
             graph['toothpaste'].sum(), graph['bathingsoap'].sum(),
             graph['shampoo'].sum(),graph['moisturizer'].sum()]

plt.figure(figsize=(10,5))
plt.pie(circledata, labels=['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')
plt.legend(loc= "lower right")
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
facew  =graph.loc[:,'facewash']
baths   = graph.loc[:,'bathingsoap']
months = graph.loc[:,'month_number']


s,subplt = plt.subplots(2, sharex=True)
subplt[0].plot(months, baths, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
subplt[0].set_title('Sales data of  a Bathingsoap')
subplt[1].plot(months, facew, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
subplt[1].set_title('Sales data of  a facewash')


plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
facc  =graph.loc[:,'facecream']
facw   = graph.loc[:,'facewash']
totp = graph.loc[:,'toothpaste']
bats   = graph.loc[:,'bathingsoap']
shp  = graph.loc[:,'shampoo']
mois = graph.loc[:,'moisturizer']
months = graph.loc[:,'month_number']

plt.figure(figsize=(10,5))
plt.plot([],[],color='m', label='face Cream')
plt.plot([],[],color='c', label='Face wash')
plt.plot([],[],color='r', label='Tooth paste')
plt.plot([],[],color='k', label='Bathing soap')
plt.plot([],[],color='g', label='Shampoo')
plt.plot([],[],color='y', label='Moisturizer')

plt.stackplot(months, facc, facw, totp, bats, shp, mois, colors=['m','c','r','k','g','y'])
plt.legend(loc="upper left")
plt.xlabel('Month Number')
plt.ylabel('Sale units')
plt.show()
```
